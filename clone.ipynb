```python
import os
import cv2
import albumentations as A
import matplotlib.pyplot as plt
import random

image_dir = 'C:\\Users\\User\\Desktop\\yolo_dataset\\splits\\fold_1\\train\\images'
output_dir = 'C:\\Users\\User\\Desktop\\yolo_dataset\\splits\\fold_5'

# преобразования
transform_motion_blur = A.MotionBlur(p=1, blur_limit=(7, 7)) # размытие в движении
transform_brightness = A.RandomBrightnessContrast(brightness_limit=0.8, contrast_limit=0.2, p=1)  # увеличение яркости
transform_rotate = A.Rotate(limit=45, p=0.6) # поворот

image_files = [f for f in os.listdir(image_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

def apply_random_transform(image_path):
    image = cv2.imread(image_path)
    if image is None:
        print(f"не удалось прочитать {image_path}")
        return None
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    random_choice = random.randint(0, 2)
    if random_choice == 0:
        transformed = transform_motion_blur(image=image)
    elif random_choice == 1:
        transformed = transform_brightness(image=image)
    else:
        transformed = transform_rotate(image=image)

    return transformed['image']

def imshow(img):
    plt.imshow(img)
    plt.axis('off')
    plt.show()

for image_file in image_files[10:]:
    image_path = os.path.join(image_dir, image_file)
    augmented_image = apply_random_transform(image_path)
    output_path = os.path.join(output_dir, image_file)
    cv2.imwrite(output_path, augmented_image)
```
